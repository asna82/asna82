```python
#ADDING FLICKER NOISE
import numpy as np
from PIL import Image
import os

# Function to add flicker noise
def add_flicker_noise(image, frequency=0.1, amplitude=0.1):
  """Adds flicker noise to an image.

  Args:
    image: The input image as a NumPy array.
    frequency: The frequency of the flicker noise.
    amplitude: The amplitude of the flicker noise.

  Returns:
    The image with flicker noise added.
  """
  # Generate flicker noise
  noise = np.random.rand(*image.shape) < frequency
  noise = noise.astype(np.float32) * amplitude

  # Add flicker noise to image
  noisy_image = image + noise
  noisy_image = np.clip(noisy_image, 0, 1)  # Clip values to [0, 1]

  return noisy_image

# Path to the Gaussian noisy images
input_folder = '/content/noisy_images/'  # Update if needed
output_folder = '/content/flicker_noisy_images/'  # Folder to save flicker noisy images
os.makedirs(output_folder, exist_ok=True)

# Process Gaussian noisy images and add flicker noise
image_files = [f for f in os.listdir(input_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

for img_file in image_files:
    try:
        # Load the Gaussian noisy image
        img_path = os.path.join(input_folder, img_file)
        img = Image.open(img_path)
        img_array = np.array(img) / 255.0  # Normalize pixel values to [0, 1]

        # Add flicker noise
        noisy_img_array = add_flicker_noise(img_array, frequency=0.1, amplitude=0.1)

        # Convert back to image format and save
        noisy_img = (noisy_img_array * 255).astype(np.uint8)
        noisy_img = Image.fromarray(noisy_img)
        noisy_img.save(os.path.join(output_folder, f"flicker_noisy_{img_file}"))

        print(f"Processed and saved flicker noisy image: flicker_noisy_{img_file}")

    except Exception as e:
        print(f"Failed to process image {img_file}: {e}")

print(f"Flicker noisy images saved to: {output_folder}")
```

```python
import cv2
import numpy as np
import os

def add_fixed_pattern_noise(image, gain_std=0.05, offset_std=5):
  """Adds fixed pattern noise to an image.

  Args:
    image: The input image as a NumPy array.
    gain_std: Standard deviation for gain variation.
    offset_std: Standard deviation for offset variation.

  Returns:
    The image with fixed pattern noise added.
  """
  # Generate gain and offset variations
  gain_map = 1 + np.random.normal(scale=gain_std, size=image.shape)
  offset_map = np.random.normal(scale=offset_std, size=image.shape)

  # Apply FPN to the image
  noisy_image = image * gain_map + offset_map
  noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)  # Clip and convert to uint8

  return noisy_image

# Path to the thermal noisy images
input_folder = '/content/thermal_noisy_images/'
# Path to save the FPN noisy images
output_folder = '/content/fpn_noisy_images/'
os.makedirs(output_folder, exist_ok=True)

# Get the list of image files
image_files = [f for f in os.listdir(input_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Process images and display
for img_file in image_files:
  try:
    # Load the thermal noisy image
    thermal_img = cv2.imread(os.path.join(input_folder, img_file))

    # Add fixed pattern noise
    fpn_img = add_fixed_pattern_noise(thermal_img, gain_std=0.05, offset_std=5)

    # Save the FPN noisy image
    cv2.imwrite(os.path.join(output_folder, f"fpn_noisy_{img_file}"), fpn_img)

    # Display images side-by-side for comparison using matplotlib.pyplot
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

    axes[0].imshow(cv2.cvtColor(thermal_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
    axes[0].set_title("Thermal Noisy Image")
    axes[0].axis('off')

    axes[1].imshow(cv2.cvtColor(fpn_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
    axes[1].set_title("FPN Noisy Image")
    axes[1].axis('off')

    plt.show()

  except Exception as e:
    print(f"Failed to process or display image {img_file}: {e}")

print("Image comparison completed.")
```

```python
import cv2
import numpy as np
import os

def add_background_fluctuation_noise(image, amplitude=10):
  """Adds background fluctuation noise to an image.

  Args:
    image: The input image as a NumPy array.
    amplitude: The amplitude of the fluctuation.

  Returns:
    The image with background fluctuation noise added.
  """
  # Generate a random background fluctuation
  fluctuation = np.random.rand(*image.shape[:2]) * amplitude
  fluctuation = fluctuation[:, :, np.newaxis]  # Add a channel dimension

  # Add fluctuation to the image
  noisy_image = image + fluctuation.astype(np.uint8)
  noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)  # Clip and convert to uint8

  return noisy_image

# Path to the FPN noisy images (input)
input_folder = '/content/fpn_noisy_images/'
# Path to save the background fluctuation noisy images (output)
output_folder = '/content/bg_fluctuation_noisy_images/'
os.makedirs(output_folder, exist_ok=True)

# Get the list of image files
image_files = [f for f in os.listdir(input_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Process images and display
for img_file in image_files:
  try:
    # Load the FPN noisy image
    fpn_img = cv2.imread(os.path.join(input_folder, img_file))

    # Add background fluctuation noise
    bg_fluctuation_img = add_background_fluctuation_noise(fpn_img, amplitude=10)

    # Save the background fluctuation noisy image
    cv2.imwrite(os.path.join(output_folder, f"bg_fluctuation_noisy_{img_file}"), bg_fluctuation_img)

    # Display images side-by-side for comparison using matplotlib.pyplot
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

    axes[0].imshow(cv2.cvtColor(fpn_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
    axes[0].set_title("FPN Noisy Image")
    axes[0].axis('off')

    axes[1].imshow(cv2.cvtColor(bg_fluctuation_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
    axes[1].set_title("Background Fluctuation Noisy Image")
    axes[1].axis('off')

    plt.show()

  except Exception as e:
    print(f"Failed to process or display image {img_file}: {e}")

print("Image comparison completed.")
```

```python
import cv2
import numpy as np
import os

def add_photon_noise(image, peak=1.0):
  """Adds Poisson (photon) noise to an image.

  Args:
    image: The input image as a NumPy array (dtype=uint8).
    peak: The peak value for scaling the noise (default=1.0).

  Returns:
    The image with Poisson noise added.
  """
  # Convert image to float32 and scale for Poisson noise
  image_float = image.astype(np.float32) * peak

  # Add Poisson noise
  noisy_image = np.random.poisson(image_float) / peak

  # Clip values to 0-255 and convert back to uint8
  noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

  return noisy_image

# Path to the background fluctuation noisy images (input)
input_folder = '/content/bg_fluctuation_noisy_images/'  # Update if needed
# Path to save the photon noisy images (output)
output_folder = '/content/photon_noisy_images/'
os.makedirs(output_folder, exist_ok=True)

# Get the list of image files
image_files = [f for f in os.listdir(input_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Process images and display
for img_file in image_files:
  try:
    # Load the background fluctuation noisy image
    bg_fluctuation_img = cv2.imread(os.path.join(input_folder, img_file))

    # Add photon noise
    photon_noisy_img = add_photon_noise(bg_fluctuation_img, peak=1.0)  # Adjust peak as needed

    # Save the photon noisy image
    cv2.imwrite(os.path.join(output_folder, f"photon_noisy_{img_file}"), photon_noisy_img)

    # Display images side-by-side for comparison using matplotlib.pyplot
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

    axes[0].imshow(cv2.cvtColor(bg_fluctuation_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
    axes[0].set_title("Background Fluctuation Noisy Image")
    axes[0].axis('off')

    axes[1].imshow(cv2.cvtColor(photon_noisy_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
    axes[1].set_title("Photon Noisy Image")
    axes[1].axis('off')

    plt.show()

  except Exception as e:
    print(f"Failed to process or display image {img_file}: {e}")

print("Image comparison completed.")
```

```python
import cv2
import numpy as np
import os

def add_readout_noise(image, stddev=5):
    """Adds Gaussian (readout) noise to an image.

    Args:
        image: The input image as a NumPy array (dtype=uint8).
        stddev: The standard deviation of the Gaussian distribution (default=5).

    Returns:
        The image with readout noise added.
    """
    # Generate Gaussian noise
    noise = np.random.normal(scale=stddev, size=image.shape)

    # Add noise to the image
    noisy_image = image + noise.astype(np.uint8)  # Ensure noise is uint8

    # Clip values to 0-255 and convert back to uint8
    noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

    return noisy_image

# Path to the photon noisy images (input)
input_folder = '/content/photon_noisy_images/'  # Update if needed
# Path to save the readout noisy images (output)
output_folder = '/content/readout_noisy_images/'
os.makedirs(output_folder, exist_ok=True)

# Get the list of image files
image_files = [f for f in os.listdir(input_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Process images and display
for img_file in image_files:
    try:
        # Load the photon noisy image
        photon_noisy_img = cv2.imread(os.path.join(input_folder, img_file))

        # Add readout noise
        readout_noisy_img = add_readout_noise(photon_noisy_img, stddev=5)  # Adjust stddev as needed

        # Save the readout noisy image
        cv2.imwrite(os.path.join(output_folder, f"readout_noisy_{img_file}"), readout_noisy_img)

        # Display images side-by-side for comparison using matplotlib.pyplot
        fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

        axes[0].imshow(cv2.cvtColor(photon_noisy_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
        axes[0].set_title("Photon Noisy Image")
        axes[0].axis('off')

        axes[1].imshow(cv2.cvtColor(readout_noisy_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
        axes[1].set_title("Readout Noisy Image")
        axes[1].axis('off')

        plt.show()

    except Exception as e:
        print(f"Failed to process or display image {img_file}: {e}")

print("Image comparison completed.")
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

def add_temperature_dependent_responsivity(image, temperature, responsivity_coeff=0.01):
    """Simulates the effect of temperature-dependent responsivity on a thermal image.

    Args:
        image: The input image as a NumPy array (dtype=uint8).
        temperature: The temperature of the scene in Celsius.
        responsivity_coeff: The coefficient representing the change in
                             responsivity per degree Celsius.

    Returns:
        The image with temperature-dependent responsivity applied.
    """
    # Convert image to float32 for calculations
    image_float = image.astype(np.float32)

    # Calculate responsivity change based on temperature
    responsivity_change = responsivity_coeff * temperature

    # Apply responsivity change to the image
    adjusted_image = image_float * (1 + responsivity_change)

    # Clip values to 0-255 and convert back to uint8
    adjusted_image = np.clip(adjusted_image, 0, 255).astype(np.uint8)

    return adjusted_image

# Path to the readout noisy images (input)
input_folder = '/content/readout_noisy_images/'
# Path to save the responsivity noisy images (output)
output_folder = '/content/responsivity_noisy_images/'
os.makedirs(output_folder, exist_ok=True)

image_files = [f for f in os.listdir(input_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Set a default scene temperature (you might need to adjust this)
scene_temperature = 25  # Example temperature in Celsius

# Process images and display
for img_file in image_files:
    try:
        # Load the image
        img_path = os.path.join(input_folder, img_file)
        img = cv2.imread(img_path)

        # Add temperature-dependent responsivity
        responsivity_img = add_temperature_dependent_responsivity(img, scene_temperature)

        # Save the image with responsivity applied
        output_path = os.path.join(output_folder, f"responsivity_{img_file}")
        cv2.imwrite(output_path, responsivity_img)

        # Display the original and adjusted images
        fig, axes = plt.subplots(1, 2, figsize=(10, 5))
        axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        axes[0].set_title("Original Image")
        axes[0].axis('off')
        axes[1].imshow(cv2.cvtColor(responsivity_img, cv2.COLOR_BGR2RGB))
        axes[1].set_title("Responsivity Adjusted Image")
        axes[1].axis('off')
        plt.show()

    except Exception as e:
        print(f"Failed to process or display image {img_file}: {e}")

print("Image processing and display completed.")
```

```python
from PIL import Image
import os

def upscale_images(input_folder, output_folder, upscale_factor=2):
    """Upscales images in a folder using PIL.Image.resize().

    Args:
        input_folder: Path to the folder containing input images.
        output_folder: Path to the folder where upscaled images will be saved.
        upscale_factor: Factor by which to upscale the images (default: 2).
    """

    os.makedirs(output_folder, exist_ok=True)  # Create output folder if it doesn't exist

    for filename in os.listdir(input_folder):
        if filename.endswith(('.jpg', '.png', '.jpeg')):  # Process only image files
            img_path = os.path.join(input_folder, filename)
            img = Image.open(img_path)

            # Get original image size
            width, height = img.size

            # Calculate new size
            new_width = int(width * upscale_factor)
            new_height = int(height * upscale_factor)

            # Resize the image
            upscaled_img = img.resize((new_width, new_height), Image.BICUBIC)  # Use BICUBIC for better quality

            # Save the upscaled image
            output_path = os.path.join(output_folder, filename)
            upscaled_img.save(output_path)

            print(f"Upscaled and saved: {filename}")

# Example usage:
input_folder = '/content/responsivity_noisy_images/'  # Replace with your input folder
output_folder = '/content/upscaled_images/'  # Replace with your output folder
upscale_factor = 2  # Set the desired upscale factor

upscale_images(input_folder, output_folder, upscale_factor)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model, Input

def cascaded_residual_block(input_tensor, filters, stage_name):
    """
    A single residual block for the cascaded model.
    """
    # First convolution
    x = layers.Conv2D(filters, (3, 3), padding='same', activation='relu', name=f'{stage_name}_conv1')(input_tensor)
    x = layers.BatchNormalization(name=f'{stage_name}_bn1')(x)

    # Second convolution
    x = layers.Conv2D(filters, (3, 3), padding='same', activation=None, name=f'{stage_name}_conv2')(x)
    x = layers.BatchNormalization(name=f'{stage_name}_bn2')(x)

    # Residual connection
    x = layers.Add(name=f'{stage_name}_residual')([x, input_tensor])
    x = layers.ReLU(name=f'{stage_name}_relu')(x)

    return x

def two_stage_cascaded_residual_cnn(input_shape, num_classes):
    """
    Builds the two-stage cascaded residual CNN model.
    """
    # Input layer
    input_layer = Input(shape=input_shape, name='input_layer')

    # Stage 1: Coarse feature extraction
    stage1 = layers.Conv2D(64, (7, 7), padding='same', activation='relu', name='stage1_conv')(input_layer)
    stage1 = layers.MaxPooling2D((2, 2), name='stage1_pool')(stage1)

    # Add residual blocks to Stage 1
    stage1 = cascaded_residual_block(stage1, 64, stage_name='stage1_block1')
    stage1 = cascaded_residual_block(stage1, 64, stage_name='stage1_block2')

    # Stage 2: Feature refinement
    stage2 = layers.Conv2D(128, (3, 3), padding='same', activation='relu', name='stage2_conv')(stage1)
    stage2 = layers.MaxPooling2D((2, 2), name='stage2_pool')(stage2)

    # Add residual blocks to Stage 2
    stage2 = cascaded_residual_block(stage2, 128, stage_name='stage2_block1')
    stage2 = cascaded_residual_block(stage2, 128, stage_name='stage2_block2')

    # Global average pooling and dense layers
    gap = layers.GlobalAveragePooling2D(name='global_avg_pool')(stage2)
    dense = layers.Dense(256, activation='relu', name='dense1')(gap)
    output_layer = layers.Dense(num_classes, activation='softmax', name='output_layer')(dense)

    # Build model
    model = Model(inputs=input_layer, outputs=output_layer, name='Two_Stage_Cascaded_Residual_CNN')
    return model

# Model configuration
input_shape = (224, 224, 3)  # Example input shape for image data
num_classes = 10             # Example number of classes

# Build the model
model = two_stage_cascaded_residual_cnn(input_shape, num_classes)

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Display the model architecture
model.summary()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np
import os
import splitfolders

# 1. Prepare the Dataset:

# Split the dataset into train, validation, and test sets:
splitfolders.ratio('/content/all_images/', output='/content/split_dataset/', seed=1337, ratio=(.8, .1, .1), group_prefix=None) # default values


# Define paths to the training, validation, and test sets:
train_dir = '/content/split_dataset/train'
val_dir = '/content/split_dataset/val'
test_dir = '/content/split_dataset/test'

# Create data generators for image augmentation and loading:
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)

# Load images from directories using data generators:
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),  # Adjust image size as needed
    batch_size=32,  # Adjust batch size as needed
    class_mode='categorical'  # Use 'binary' for binary classification
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False  # Important for evaluation metrics
)


# 2. Build and Train the Model:

# Create a simple CNN model (adjust architecture as needed):
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(train_generator.num_classes, activation='softmax')  # Number of classes
])

# Compile the model:
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model:
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,  # Adjust number of epochs as needed
    validation_data=val_generator,
    validation_steps=val_generator.samples // val_generator.batch_size
)


# 3. Evaluate the Model:

# Evaluate the model on the test set:
test_loss, test_acc = model.evaluate(
    test_generator,
    steps=test_generator.samples // test_generator.batch_size
)

print(f"Test accuracy: {test_acc}")


# 4. Save the Model:

# Save the trained model:
model.save('my_trained_model.h5')

# Plot training history (optional):
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()
```

```python
# CLASSIFICATION LAYER
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set paths for training and validation datasets
train_data_dir = '/content/split_dataset/train'  # Replace with your actual path
validation_data_dir = '/content/split_dataset/val'  # Replace with your actual path

# Data Augmentation and Normalization
datagen = ImageDataGenerator(
    rescale=1./255  # Normalizes pixel values to the range [0, 1]
)

# Create data generators
train_generator = datagen.flow_from_directory(
    train_data_dir,
    target_size=(224, 224),  # Adjust target size as needed
    batch_size=32,  # Adjust batch size as needed
    class_mode='categorical'  # Or 'binary' depending on your task
)

validation_generator = datagen.flow_from_directory(
    validation_data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'  # Or 'binary'
)

# Get input shape from train_generator
input_shape = train_generator.image_shape

# Define the Model
model = tf.keras.Sequential([
    # Initial Convolutional Block
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    tf.keras.layers.MaxPooling2D((2, 2)),

    # Convolution Block with BatchNorm and ReLU
    tf.keras.layers.Conv2D(64, (3, 3)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.ReLU(),
    tf.keras.layers.MaxPooling2D((2, 2)),

    # Depthwise Separable Convolution Block
    tf.keras.layers.DepthwiseConv2D((3, 3), activation='relu', depth_multiplier=1),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(128, (1, 1), activation='relu'),  # Pointwise convolution
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),

    # Additional Convolutional Blocks
    tf.keras.layers.Conv2D(256, (3, 3)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.ReLU(),
    tf.keras.layers.MaxPooling2D((2, 2)),

    tf.keras.layers.Conv2D(512, (3, 3)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.ReLU(),
    tf.keras.layers.GlobalAveragePooling2D(),

    # Classification Layer with Softmax
    tf.keras.layers.Dense(4, activation='softmax')  # Adjust '4' based on your number of classes
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=validation_generator
)

# Plot training accuracy and loss
plt.figure(figsize=(12, 5))

# Accuracy Plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss Plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()

# Save the model (optional)
model.save('/content/classification_model.h5')
```
